```python
import jcamp
import glob
import os
from collections import Counter
from tqdm import tqdm

jdx_path = '/Users/benji/phd/data/phd/attentivefp/jdx'
# read all jdx files in the jdx directory
# make counter dict of the yunits and xunits
yunits_counter = Counter()
xunits_counter = Counter()
xfactors_counter = Counter()
yfactors_counter = Counter()
state_counter = Counter()
for jdx_path in tqdm(glob.glob(os.path.join(jdx_path, '*.jdx')), desc='Processing JDX files', total=len(glob.glob(os.path.join(jdx_path, '*.jdx')))):
    try:
        jdx_data = jcamp.jcamp_readfile(jdx_path)
        
        # Check if data is absorbance and convert to transmittance if needed
        y_units = jdx_data.get('yunits', '').upper()
        if y_units not in yunits_counter:
            yunits_counter[y_units] = 1
        else:
            yunits_counter[y_units] += 1
        y_factor = jdx_data.get('yfactor', 1)
        
        # convert xunits to wavenumbers
        x_units = jdx_data.get('xunits', '').upper()
        if x_units not in xunits_counter:
            xunits_counter[x_units] = 1
        else:
            xunits_counter[x_units] += 1
        x_factor = jdx_data.get('xfactor', 1)
        if x_factor not in xfactors_counter:
            xfactors_counter[x_factor] = 1
        else:
            xfactors_counter[x_factor] += 1
        if y_factor not in yfactors_counter:
            yfactors_counter[y_factor] = 1
        else:
            yfactors_counter[y_factor] += 1
        
        state = jdx_data.get('state', '')
        if state not in state_counter:
            state_counter[state] = 1
        else:
            state_counter[state] += 1
        
    except Exception as e:
        print(f"Error processing {jdx_path}: {str(e)}")

print(yunits_counter)
print(xunits_counter)
```
